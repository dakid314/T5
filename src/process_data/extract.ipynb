```python
from Bio import Entrez, SeqIO
import pandas as pd

data = pd.read_csv('data/prot_name.csv')
Entrez.email = 'dakid314@163.com'
# 定义蛋白质编号列表
protein_ids = list(data['prot'])
```

```python
protein_ids
```

```python
for protein_id in protein_ids:
    if  Entrez.esearch(db="protein", term=protein_id)==True:
        handle = Entrez.efetch(db='protein', id=protein_id, rettype='fasta', retmode='text')
        record = SeqIO.read(handle, 'fasta')
        SeqIO.write(record, f'data/f_data/{protein_id}.fasta', 'fasta')
        handle.close()
```

```python
def filter_fasta(input_file, output_file_with, output_file_without, specific_char):
    with open(input_file, 'r') as f_in, open(output_file_with, 'w') as f_out_with, open(output_file_without, 'w') as f_out_without:

        proteins = []
        current_protein_annotation = ""
        current_protein_sequence = ""

        
        for line in f_in:
            line = line.strip()
            if line.startswith('>'):
                if current_protein_annotation != "":
                    proteins.append((current_protein_annotation, current_protein_sequence))
                    current_protein_annotation = ""
                    current_protein_sequence = ""
                current_protein_annotation = line[1:]
            else:
                current_protein_sequence += line

    # 处理最后一条蛋白质序列
        if current_protein_annotation != "":
            proteins.append((current_protein_annotation, current_protein_sequence))

        for i, (annotation, sequence) in enumerate(proteins):
            print(f"蛋白质 {i+1} 的注释行:", annotation)
            print(f"蛋白质 {i+1} 的序列:", sequence)
            if specific_char in annotation:
                f_out_with.write(f'>{annotation}\n')
                f_out_with.write(f'{sequence}\n')
                    
            else:
                f_out_without.write(f'>{annotation}\n')
                f_out_without.write(f'{sequence}\n')
filter_fasta('data/T5_70.fasta', 'output_with.fasta', 'output_without.fasta', 'Escherichia coli')
```
